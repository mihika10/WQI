```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Assume 'data' is your preprocessed dataset with 30 columns

# Perform PCA
pca = PCA(n_components=0.9)  # Retain components explaining 90% of the variance
reduced_data = pca.fit_transform(data)

# Analyze explained variance ratios
explained_variance_ratios = pca.explained_variance_ratio_
cumulative_variance_ratios = np.cumsum(explained_variance_ratios)

# Plot explained variance ratios
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variance_ratios) + 1), explained_variance_ratios, alpha=0.8)
plt.plot(range(1, len(cumulative_variance_ratios) + 1), cumulative_variance_ratios, color='red', marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratios')
plt.xticks(range(1, len(explained_variance_ratios) + 1))
plt.show()

# Visualize reduced-dimensional data
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.8)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Scatter Plot of Principal Components 1 and 2')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load your dataset from 'TRIAL_2.xlsx' using pandas or any other method
data = pd.read_excel('TRIAL_2.xlsx')

# Perform any necessary preprocessing steps on the data

# Perform PCA
pca = PCA(n_components=0.9)  # Retain components explaining 90% of the variance
reduced_data = pca.fit_transform(data)

# Analyze explained variance ratios
explained_variance_ratios = pca.explained_variance_ratio_
cumulative_variance_ratios = np.cumsum(explained_variance_ratios)

# Plot explained variance ratios
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variance_ratios) + 1), explained_variance_ratios, alpha=0.8)
plt.plot(range(1, len(cumulative_variance_ratios) + 1), cumulative_variance_ratios, color='red', marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratios')
plt.xticks(range(1, len(explained_variance_ratios) + 1))
plt.show()

# Visualize reduced-dimensional data
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.8)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Scatter Plot of Principal Components 1 and 2')
plt.show()
```
